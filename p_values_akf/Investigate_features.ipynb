```python
from p_value_matrix import p_value_matrix
import matplotlib.pyplot as plt
from pandas.plotting import table 
```

```python
song_features = ['Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness', 'Acousticness', 'Instramentalness',
                    'Liveness', 'Valence', 'Tempo', 'Duration (ms)', 'Time Signature']
```

```python
for feature in song_features:
    df = p_value_matrix(feature)
#     df.to_excel(f'p_value_excel/{feature}_p_values.xlsx')
    
    ax = plt.subplot(111, frame_on=False) # no visible frame
    ax.xaxis.set_visible(False)  # hide the x axis
    ax.yaxis.set_visible(False)  # hide the y axis

    table(ax, df)  # where df is your data frame

    plt.savefig(f'p_value_excel/{feature}_p_values.png')
```
